## 1. Limits

We say that the limit of $ f(x) $ is $L$ as x approaches a and write this as

$\lim_{x \to a} f(x) = L$

we can make $f(x)$ as close to $L$ as we want for all x sufficiently close to a, from both sides, without actually letting x be h



Simple examples :
1. $\lim_{x \to 4} \sqrt{100 - x^2}$ 
57. $\lim_{x \to 2} \frac{x-2}{x^2-4}$
44. $\lim_{x \to 2} \frac{1}{x-2}$
3. $\lim_{x \to 2} log(x-2)$

```python
# domain + functions
x = np.linspace(8, 12, 400)
y1 = np.sqrt(13 - x**2)
y2 = (x - 2)/x**2 - 4
y3 = 1/(x - 2)
y4 = np.log(x - 2)

funcs = [(y1, r"$\sqrt{25 - x^2}$"),
         (y2, r"$(x-2)/x^2 - 4$"),
         (y3, r"$1/(x-2)$"),
         (y4, r"$\ln(x-2)$")]

a = 2 

fig, axes = plt.subplots(2, 2, figsize=(10, 8), sharex=True)
for ax, (y, label) in zip(axes.flat, funcs):
    ax.plot(x, y, label=label)
    ax.axvline(a, linestyle="--")  # brute-force vertical line
    ax.set_title(label)
    ax.grid(True)

for ax in axes[-1]: ax.set_xlabel("x")
for ax in axes[:,0]: ax.set_ylabel("y")

plt.tight_layout()
plt.show()

```

## 2. Derivatives

```python
f = lambda x: x**2
x = np.linspace(-2, 2, 400)
y = f(x)

x0 = 1
h_values = [10, 5, 0.1, 0.01]

plt.plot(x, y, label="$f(x) = x^2$")
plt.scatter([x0], [f(x0)], color="red")

for h in h_values:
    slope = (f(x0+h) - f(x0)) / h
    tangent_line = slope*(x - x0) + f(x0)
    plt.plot(x, tangent_line, linestyle="--", label=f"h={h}")

plt.legend()
plt.title("Derivative as Limit of Secant Slopes")
plt.show()
```
